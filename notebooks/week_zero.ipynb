```python
#imports
import pandas as pd
import os
```

```python
CSV_DIR = "../data/csv"
DIR = os.listdir(CSV_DIR)

sold_dfs = []
listings_dfs = []

for file in DIR:
    if "Sold" in file:
        temp_df = pd.read_csv(os.path.join(CSV_DIR, file))
        temp_df.name = file
        sold_dfs.append(temp_df)
    elif "Listing" in file:
        temp_df = pd.read_csv(os.path.join(CSV_DIR, file))
        temp_df.name = file
        listings_dfs.append(temp_df)
```

```python
sold_df = pd.concat(sold_dfs, ignore_index=True)
listings_df = pd.concat(listings_dfs, ignore_index=True)
```
